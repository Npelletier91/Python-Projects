```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
```

```python
norm_dis = np.random.normal(loc=0, scale= 1, size=100000)
norm_dis[:10]
```

```python
sample_mean = np.mean(norm_dis)
sample_mean
```

```python
sample_med = np.median(norm_dis)
sample_med
```

```python
sample_std = np.std(norm_dis)
sample_std
```

```python
plt.figure(figsize=(10,6))
sns.histplot(x=norm_dis)
plt.axvline(x=sample_mean)
plt.axvline(x=sample_med)

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(x=norm_dis)
plt.axvline(x=sample_mean)
plt.axvline(x=sample_med)

plt.show()
```

```python
norm_dis1 = np.random.normal(loc=2, scale= 1, size=100000)
norm_dis2 = np.random.normal(loc=0, scale= 3, size=100000)


plt.figure(figsize=(10,6))

sns.kdeplot(x=norm_dis)
sns.kdeplot(x=norm_dis1)
sns.kdeplot(x=norm_dis2)

plt.axvline(x=sample_mean)
plt.axvline(x=sample_med)

plt.show()
```
